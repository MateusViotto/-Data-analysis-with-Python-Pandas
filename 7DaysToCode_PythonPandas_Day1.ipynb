## Imports pandas and numpy

```python
import pandas as pd
import numpy as np
```

## Reads datasets and concatenates into a table

```python
list_advance_temp = []
raw_default = "https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-20"
for i in range(10):
    for k in range(2):
        raw_current = raw_default + str(int(10+i)) + str(k+1) + ".csv"
        table_temp = pd.read_csv(raw_current)
        list_advance_temp.append(table_temp)
        
raw_current = raw_default + "201.csv"
table_temp = pd.read_csv(raw_current)
list_advance_temp.append(table_temp)

table_advance = pd.concat(list_advance_temp, ignore_index=True)
display(table_advance)
```

## Reads additional dataset

```python
collection = pd.read_parquet("https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet?raw=true")
display(collection)
table_advance = table_advance.drop_duplicates()
```

## Merges tables

```python
df = table_advance.merge(collection)
df
```

## Removes registro_sistema

```python
df = df.drop(columns = ['registro_sistema'])
df
```

## Fixes matricula_ou_siape

```python
df['matricula_ou_siape'] = df['matricula_ou_siape'].astype('string')
display(df)
```

## Creates classe_geral

```python
conditions = [    (df['localizacao'] >= 0) & (df['localizacao'] <= 99),
    (df['localizacao'] <= 199) & (df['localizacao'] >= 100),
    (df['localizacao'] <= 299) & (df['localizacao'] >= 200),
      (df['localizacao'] <= 399) & (df['localizacao'] >= 300),
      (df['localizacao'] <= 499) & (df['localizacao'] >= 400),
      (df['localizacao'] <= 599) & (df['localizacao'] >= 500),
      (df['localizacao'] <= 699) & (df['localizacao'] >= 600),
      (df['localizacao'] <= 799) & (df['localizacao'] >= 700),
      (df['localizacao'] <= 899) & (df['localizacao'] >= 800),
      (df['localizacao'] <= 999) & (df['localizacao'] >= 900)
]

values = ['Generalidades. Ciência e conhecimento', 'Filosofia e psicologia', 'Religião', 'Ciências sociais',
           'Classe vaga. Provisoriamente não ocupada', 'Matemática e ciências naturais', 'Ciências aplicadas',
           'Belas artes', 'Linguagem. Língua. Linguística', 'Geografia. Biografia. História'
]

df['classe_geral'] = np.select(conditions,values)
df
```

## Imports matplotlib

```python
import matplotlib.pyplot as plt
```

## Plots yearly lends

```python
df_books_year = df[['data_emprestimo','id_exemplar']].copy()
df_books_year['data_emprestimo'] = pd.to_datetime(df_books_year['data_emprestimo'])
df_books_year['data_emprestimo'] = df_books_year['data_emprestimo'].dt.year
df_books_year = df_books_year.groupby('data_emprestimo', as_index=False)['id_exemplar'].nunique()


plt.plot(df_books_year['data_emprestimo'], df_books_year['id_exemplar'], color = 'red', marker = 'o')
plt.title('Lended copies by year')
plt.xlabel('Year')
plt.ylabel('Number of lended copies')
plt.show()
```

## Plots monthly lends

```python
df_books_month = df[['data_emprestimo','id_exemplar']].copy()
df_books_month['data_emprestimo'] = pd.to_datetime(df_books_month['data_emprestimo'])
df_books_month['data_emprestimo'] = df_books_month['data_emprestimo'].dt.month
df_books_month = df_books_month.groupby('data_emprestimo', as_index=False)['id_exemplar'].nunique()

plt.plot(df_books_month['data_emprestimo'], df_books_month['id_exemplar'], color = 'green', marker = 'o')
plt.title('Lended copies by month')
plt.xlabel('Month')
plt.ylabel('Number of lended copies')
plt.xticks(np.arange(0, 12.5, 1))
plt.show()
```

## Plots hourly lends

```python
df_books_hour = df[['data_emprestimo','id_exemplar']].copy()
df_books_hour['data_emprestimo'] = pd.to_datetime(df_books_hour['data_emprestimo'])
df_books_hour['data_emprestimo'] = df_books_hour['data_emprestimo'].dt.hour

list_hours = range(24)
list_books_hour = []
for i in range(24):
    list_books_hour.append((df_books_hour['data_emprestimo'] == i).sum())

plt.bar(list_hours, list_books_hour, color = 'orange')
plt.title('Lends by hour')
plt.xlabel('Hour')
plt.ylabel('Number of lends')
plt.xticks(np.arange(0, 24, 1))
plt.show()
```

## Function for frequency analysis

```python
def tableFrequency(column):
    df_frequency = pd.DataFrame()
    df_frequency['frequencia_absoluta'] = df[column].value_counts()
    total = df_frequency['frequencia_absoluta'].sum()
    df_frequency['frequencia_relativa'] = round((df_frequency['frequencia_absoluta']/total * 100).astype(float), 2)
    return df_frequency
```

## Filters necessary data and creates database

```python
df_base = df.loc[(df['classe_geral'] == 'Ciências aplicadas') & ((df['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO') | 
(df['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO')), ['data_emprestimo', 'tipo_vinculo_usuario']]
df_base['data_emprestimo'] = pd.to_datetime(df_base['data_emprestimo'])

df_graduacao = []
df_posgrad = []
for i in range(11):
    df_graduacao.append(df_base.loc[(df_base['data_emprestimo'].dt.year == (2010 + i))
& (df_base['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO'), ['data_emprestimo']])
    df_posgrad.append(df_base.loc[(df_base['data_emprestimo'].dt.year == (2010 + i))
& (df_base['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO'), ['data_emprestimo']])
for k in range(11):
    df_graduacao[k]['data_emprestimo'] = df_graduacao[k]['data_emprestimo'].dt.month
    df_posgrad[k]['data_emprestimo'] = df_posgrad[k]['data_emprestimo'].dt.month
```

## Plots graphs

```python
def YearlyLoans(year):
    df_graduacao[year-2010]['data_emprestimo'].value_counts().sort_index().plot(color='red')
    plt.title('Empréstimos em {} por alunos de graduação'.format(year))
    plt.xlabel('Mês')
    plt.ylabel('Quantidade de empréstimos')
    plt.xticks(np.arange(1, 13, 1))
    plt.show()
    df_posgrad[year-2010]['data_emprestimo'].value_counts().sort_index().plot(color='green')
    plt.title('Empréstimos em {} por alunos de pós-graduação'.format(year))
    plt.xlabel('Mês')
    plt.ylabel('Quantidade de empréstimos')
    plt.xticks(np.arange(1, 13, 1))
    plt.show()
```
