```python
import pandas as pd
import numpy as np
```

## Reads datasets and concatenates into a table

```python
list_advance_temp = []
raw_default = "https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-20"
for i in range(10):
    for k in range(2):
        raw_current = raw_default + str(int(10+i)) + str(k+1) + ".csv"
        table_temp = pd.read_csv(raw_current)
        list_advance_temp.append(table_temp)
        
raw_current = raw_default + "201.csv"
table_temp = pd.read_csv(raw_current)
list_advance_temp.append(table_temp)

table_advance = pd.concat(list_advance_temp, ignore_index=True)
display(table_advance)
```

## Reads additional dataset

```python
collection = pd.read_parquet("https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet?raw=true")
display(collection)
table_advance = table_advance.drop_duplicates()
```

## Merges tables

```python
table_advance = table_advance.merge(collection)
table_advance
```

## Removes registro_sistema

```python
table_advance = table_advance.drop(columns = ['registro_sistema'])
table_advance
```

## Fixes matricula_ou_siape

```python
table_advance['matricula_ou_siape'] = table_advance['matricula_ou_siape'].astype('string')
display(table_advance)
```

## Creates classe_geral

```python
conditions = [    (table_advance['localizacao'] >= 0) & (table_advance['localizacao'] <= 99),
    (table_advance['localizacao'] <= 199) & (table_advance['localizacao'] >= 100),
    (table_advance['localizacao'] <= 299) & (table_advance['localizacao'] >= 200),
      (table_advance['localizacao'] <= 399) & (table_advance['localizacao'] >= 300),
      (table_advance['localizacao'] <= 499) & (table_advance['localizacao'] >= 400),
      (table_advance['localizacao'] <= 599) & (table_advance['localizacao'] >= 500),
      (table_advance['localizacao'] <= 699) & (table_advance['localizacao'] >= 600),
      (table_advance['localizacao'] <= 799) & (table_advance['localizacao'] >= 700),
      (table_advance['localizacao'] <= 899) & (table_advance['localizacao'] >= 800),
      (table_advance['localizacao'] <= 999) & (table_advance['localizacao'] >= 900)
]

values = ['Generalidades. Ciência e conhecimento', 'Filosofia e psicologia', 'Religião', 'Ciências sociais',
           'Classe vaga. Provisoriamente não ocupada', 'Matemática e ciências naturais', 'Ciências aplicadas',
           'Belas artes', 'Linguagem. Língua. Linguística', 'Geografia. Biografia. História'
]

table_advance['classe_geral'] = np.select(conditions,values)
table_advance
```
